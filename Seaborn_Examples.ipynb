## Distribution of scores among different income levels among improving states

**From the 1st histogram below for the math scores we observe that**

the min, mean and max of the income classes above 40k are pretty similar. This might imply that in our improving states if you are above the 40k threshold your average score will not be affected drastically by your income level.  

A very fascinating outlier lies in the 20-40k income level where we see that the max score for this income level actual exceeds all other income levels. This to me implies that there are people of varying skills at every income level and if given the chance they would rise to the top. 

**From the 2nd histogram below for the verbal scores we observe that**

there is a clear seperation of the 60K+ group and under 60k group.  This suggests to me that verbal skills, unlike mathematical skills are highly correlated with income in the form of outside classes, tutors and so forth.

## Correlation between GPA and SAT Scores

**Looking at the plot below**

The relationship between math gpa, English gpa and SAT Math scores, SAT Verbal scores is as expected there have postive trends.  

What is surprising is to see that extremly strong correlation between Arts/music grades with SAT scores.  This could have multiple confounding variables attached to it, for example it is probably a fact that wealthier families can afford music lessons which also leads to hiring tutors and getting a better score.  Further analysis will need to be done to see if such confounding variables are removed what will be the correlation but for now we will go with higher arts and music scores leads to higher SAT scores.

```python
data_Math = np.array(states_improving[['Mathematics.Average GPA','Math']])
data_Verbal = np.array(states_improving[['English.Average GPA','Verbal']])
data_Music_Verbal = np.array(states_improving[['Arts/Music.Average GPA','Verbal']])
data_Music_Math = np.array(states_improving[['Arts/Music.Average GPA','Math']])
```

```python
df = pd.DataFrame(data_Math, columns=["Math GPA", "Math Score"])

sns.jointplot(x="Math GPA", y="Math Score", data=df);


plt.show() 

sns.regplot(x="Math GPA", y="Math Score", data=df);
```

```python
df = pd.DataFrame(data_Verbal, columns=["English GPA", "Verbal Score"])

sns.jointplot(x="English GPA", y="Verbal Score", data=df);


plt.show() 
sns.regplot(x="English GPA", y="Verbal Score", data=df);
```

```python
df = pd.DataFrame(data_Music_Math, columns=["Arts/music GPA", "Math"])

sns.jointplot(x="Arts/music GPA", y="Math", data=df);


plt.show()


sns.regplot(x="Arts/music GPA", y="Math", data=df);
```

```python
df = pd.DataFrame(data_Music_Verbal, columns=["Arts/Music GPA", "Verbal"])

sns.jointplot(x="Arts/Music GPA", y="Verbal",data=df);


plt.show() 

sns.regplot(x="Arts/Music GPA", y="Verbal",data=df);
```

## Does Gender effect the correlation between Arts/Music GPA and SAT Score?

Examining our swarm plots below except a few a outliers it is clear that Arts/Music GPA effects both female and males similarly in terms of correlation between gpa and SAT scores.  One thing that is clear though is that females **consistantly have higher overall scores that their male counter parts**.   This is very interesting when you think about the job market and inequlities of pay for male and females and can lead to further studies.
